## 12.(0,5 ponto) Quais são as cidades e respectivas populações com menor e maior número de habitantes no dataset?

```python
# Encontrando o índice da cidade com a maior população
id_maior_populacao = df['populacao'].idxmax()

# Recuperando o nome da cidade usando o índice e armazenando a sua respectiva população
cidade_maior_populacao = df.loc[id_maior_populacao, 'municipio']
num_cidade_maior_populacao = df.loc[id_maior_populacao, 'populacao']

# Encontrando o índice da cidade com a menor população
id_menor_populacao = df['populacao'].idxmin()

# Recuperando o nome da cidade usando o índice e armazenando a sua respectiva população
cidade_menor_populacao = df.loc[id_menor_populacao, 'municipio']
num_cidade_menor_populacao = df.loc[id_menor_populacao, 'populacao']

print("Cidade com maior população: ", cidade_maior_populacao, "(", num_cidade_maior_populacao, " habitantes )")
print("Cidade com menor população: ", cidade_menor_populacao, "(", num_cidade_menor_populacao, " habitantes )")
```

## 13.(0,5 ponto) Quais os estados com menor e maior população?

```python
# Agrupando os dados por estado e somando a população de cada estado
populacao_por_estado = df.groupby('uf')['populacao'].sum()

# Encontrando o índice do estado com a maior população
id_maior_populacao = populacao_por_estado.idxmax()

# Recuperando o nome do estado usando o índice e armazenando a sua respectiva população
uf_maior_populacao = id_maior_populacao
num_uf_maior_populacao = populacao_por_estado.loc[id_maior_populacao]

# Encontrando o índice do estado com a menor população
id_menor_populacao = populacao_por_estado.idxmin()

# Recuperando o nome do estado usando o índice e armazenando a sua respectiva população
uf_menor_populacao = id_menor_populacao
num_uf_menor_populacao = populacao_por_estado.loc[id_menor_populacao]

print("Estado com maior população: ", uf_maior_populacao, "(", num_uf_maior_populacao, " habitantes )")
print("Estado com menor população: ", uf_menor_populacao, "(", num_uf_menor_populacao, " habitantes )")
```

### a)(0,5 ponto) Acurácia:

```python
# TP (Verdadeiro Positivo): 2564
# FP (Falso Positivo): 314
# TN (Verdadeiro Negativo): 2156
# FN (Falso Negativo): 124
"""
Acurácia = (TP + TN) / (TP + FP + TN + FN) = (2564 + 2156) / (2564 + 314 + 124 + 2156) = 0.91 (91%)
"""
```

### b)(0,5 ponto) Precisão:

```python
# TP (Verdadeiro Positivo): 2564
# FP (Falso Positivo): 314
# TN (Verdadeiro Negativo): 2156
# FN (Falso Negativo): 124
"""
Precisão = TP / (TP + FP) = 2564 / (2564 + 314) = 0.89 (89%)
"""
```

### c)(0,5 ponto) Sensibilidade (Recall):

```python
# TP (Verdadeiro Positivo): 2564
# FP (Falso Positivo): 314
# TN (Verdadeiro Negativo): 2156
# FN (Falso Negativo): 124
"""
Sensibilidade = TP / (TP + FN) = 2564 / (2564 + 124) = 0.95 (95%)
"""
```

### d)(0,5 ponto) F1 - Score:

```python
# TP (Verdadeiro Positivo): 2564
# FP (Falso Positivo): 314
# TN (Verdadeiro Negativo): 2156
# FN (Falso Negativo): 124
"""
F1-Score = 2 * (Precisão * Sensibilidade) / (Precisão + Sensibilidade) = 2 * (0.89 * 0.95) / (0.89 + 0.95) = 0.91 (91%)
"""
```
